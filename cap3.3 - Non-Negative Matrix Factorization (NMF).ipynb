# Non-Negative Matrix Factorization (NMF)

Unsupervised learning algorithm for feature extraction. It works like PCA and can be used to dimensionality reduction.

Unlike PCA, NMF uses a random initialization, it can only be applied to non-nagative features, produces components that are more interpretable than PCA's components and the components are equal in importance.

```python
import mglearn

mglearn.plots.plot_nmf_illustration();
```

Removing one component, creates a different set of components.

# Applying NMF to face images

### Here we'll apply NMF to Labeled Faces in the Wild dataset. Below, we show a sample of it.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people

people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape

fix, axes = plt.subplots(2, 5, figsize=(15,8), subplot_kw={'xticks': (), 'yticks': ()})

for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
print("people.images.shape: {}".format(people.images.shape))
print("Number of classses: {}".format(len(people.target_names)))
```

## Preprocessing step (Fix 50 faces per class and transform grayscale so the values are between 0 and 1)

```python
import numpy as np

# Sample each class with 50 examples.
mask = np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1

X_people = people.data[mask]
y_people = people.target[mask]

# scale the grayscale values to be between 0 and 1
# instead of 0 and 255 for better numeric stability
X_people = X_people / 255.
```

## Separate data in train/test datasets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_people, y_people, stratify=y_people, random_state=0)
```

## Apply NMF and show the results (15 components)

```python
from sklearn.decomposition import NMF

nmf = NMF(n_components=15, random_state=0)
nmf.fit(X_train)
X_train_nmf = nmf.transform(X_train)
X_test_nmf = nmf.transform(X_test)

fix, axes = plt.subplots(3, 5, figsize=(15, 12), subplot_kw={'xticks': (), 'yticks': ()})

for i, (component, ax) in enumerate(zip(nmf.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape), cmap='viridis')
    ax.set_title("{}. component".format((i)))
```

NMF tries to explain the dataset using a linear combination of N components. In this case, we have 15 components (shown above ). Also, NMF's results are more interpretable than PCA. See how the component 3 seems like a face rotated to the right.

##  Meaning from components

```python
print('sort by 3rd component, plot first 10 images')
compn = 3
# sort by 3rd component, plot first 10 images
inds = np.argsort(X_train_nmf[:, compn])[::-1]
fig, axes = plt.subplots(2, 5, figsize=(15, 8), subplot_kw={'xticks': (), 'yticks': ()})

for i, (ind, ax) in enumerate(zip(inds, axes.ravel())):
    ax.imshow(X_train[ind].reshape(image_shape))
```

Here we can see that faces with a high coefficient for component 3 are rotated to the right.

```python
print('sort by 7rd component, plot first 10 images')
compn = 7
# sort by 7th component, plot first 10 images
inds = np.argsort(X_train_nmf[:, compn])[::-1]
fig, axes = plt.subplots(2, 5, figsize=(15, 8), subplot_kw={'xticks': (), 'yticks': ()})

for i, (ind, ax) in enumerate(zip(inds, axes.ravel())):
    ax.imshow(X_train[ind].reshape(image_shape))
```

Otherwise, a high coefficient for component 7 indicates a face rotated to the left.

# NMF applied to other domains

NMF works very well when applied to data with additive structure, like audio, gene expression, and text data. Let's see an example.

```python
S = mglearn.datasets.make_signals()
plt.figure(figsize=(6, 1))
plt.plot(S, '-')
plt.xlabel("Time")
plt.ylabel("Signal");
```

```python
# mix data into a 100-dimensional state
A = np.random.RandomState(0).uniform(size=(100, 3))
X = np.dot(S, A.T)
print("Shape of measurements: {}".format(X.shape))
```

```python
nmf = NMF(n_components=3, random_state=42)
S_ = nmf.fit_transform(X)
print("Recovered signal shape: {}".format(S_.shape))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
H = pca.fit_transform(X)
```

```python
models = [X, S, S_, H]
names = ['Observations (first three measurements)', 'True sources', 'NMF recovered signals', 'PCA recovered signals']

fig, axes = plt.subplots(4, figsize=(8, 4), gridspec_kw={'hspace': .5}, subplot_kw={'xticks': (), 'yticks': ()})

for model, name, ax in zip(models, names, axes):
    ax.set_title(name)
    ax.plot(model[:, :3], '-')
```

As we can see, NMF does a great job in recovering the original signals using 3 components, while PCA fails in this task.
