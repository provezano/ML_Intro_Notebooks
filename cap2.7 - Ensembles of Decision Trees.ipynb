# Ensembles

Ensembles are combinations of several ML models.
Here, we'll overview Random Forest and Gradient Boosted Regression Trees.

# 1.0 Random Forest

Random forests solves the overfitting problem faced with decision trees. A random forest is the combination of several
(different) decision trees to perform predicions. While each decision tree faces its own overfitting issue, the mean among
several trees smooths this effect.

Here we have two main parameters n_estimators (number of decision trees) and max_features (number of features used by each decision tree).

The randomness occurs in the feature selection (if max_features < # of features) and data sample with replacement (bootstrap).

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
forest = RandomForestClassifier(n_estimators=5, random_state=2)
forest.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
import mglearn

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), forest.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(forest, X_train, fill=True, ax=axes[-1, -1], alpha=.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train);
```

The image above is every tree used to induced our Random Forest. We have 5 decision trees and a final Random Forest result.

## 1.1 Random Forest applied to the Cancer dataset

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

forest = RandomForestClassifier(n_estimators=100, random_state=0)
forest.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(forest.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(forest.score(X_test, y_test)))
```

A very good result even without complexity control.

We can improve further with parameter control,

```python
forest = RandomForestClassifier(n_estimators=100, max_depth=7, min_samples_split=6, min_samples_leaf=4, random_state=0)
forest.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(forest.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(forest.score(X_test, y_test)))
```

```python
import numpy as np
plt.rcParams['figure.figsize'] = [12, 8]

def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
```

We also have feature importances for random forests. In this case, the importances are in agreggation of importances among the decision trees.

```python
plot_feature_importances_cancer(forest)
```

Random forests tend to distribute importances among the featres. This an effect of multiple explanations given by multiple trees.
A random forest is forced to consider several view points about the fenomena we are trying to explain (predict).

# 2.0 Random Forests - Strengths, weaknesses, and parameters.

Random forests are among state-of-art ML algo. for classification and regression.
They are robust to feature scaling and parameter control. This means they tend to presend good results even without scaling the dataset and setting the correct parameters.

While the isolated decision tree is very easy to explain and visually friendly. Random Forests are harder to visualize.

Usually, random forests do not perform well on high dimensional, sparse data. The training phase can be very slow and the model very large.

n_estimators, max_features, and max_leaf_nodes are the most important parameters.

Usually, a good rule of thumb is to use max_features=sqrt(n_features) for classification and max_fea tures=log2(n_features) for regression.
And, the more n_estimators, the better.

# 3.0 Gradient Boosted Regression Trees

An ensemble method used for classification and regression, that combines several decision trees. The trees are built in a serial manner in which each tree tries to correct the mistakes of the previous one.

Here, a strong pre-prunning (max_depth ranges between 1 to 5) is used in place of randomness.

The main idea behind gradient boosting is to combine week learners (simple models or shallow trees) to improve prediction performance.

They are more sensible to parameter tuning, but usually presents better performance than Random Forests.

Here, we have parameters for the number of trees, pre-pruning and learning rate (how strongly each tree can make corrections to the previous trees.)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import GradientBoostingClassifier

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

gbrt = GradientBoostingClassifier(n_estimators = 100, max_depth = 3, learning_rate = 0.1, random_state=0)
gbrt.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(n_estimators = 100, max_depth = 1, learning_rate = 0.1, random_state=0)
gbrt.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(n_estimators = 100, max_depth = 3, learning_rate = 0.01, random_state=0)
gbrt.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
gbrt = GradientBoostingClassifier(n_estimators = 20, max_depth = 3, learning_rate = 1, random_state=0)
gbrt.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

And... Feature importances...

```python
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)
plot_feature_importances_cancer(gbrt)
```

## 3.1 Strengths, weaknesses, and parameters

Gradient boosted decision trees are among the state-of-the-art models for supervised learning.

The main drawback is that it requires a careful fine tuning of the parameters that may require a long time.

GBMs work well without feature scaling and with a mixture of binary and continuous features. But, it doesn`t work well with high dimensional sparse data.

The main parameters are: n_estimators and learning_rate which control the number of trees and the degree each tree is allowed to correct the mistakes of the previous trees.

Contrary to Random Forests, increasing n_estimators can lead to overfitting once we increase the model complexity.

Usually we set n_estimators according to the time and memory budget, and then search over different learning_rates.

Another parameter is max_depth (or max_leaf_nodes), to reduce the complexity of each tree (usually not deeper than five splits).
