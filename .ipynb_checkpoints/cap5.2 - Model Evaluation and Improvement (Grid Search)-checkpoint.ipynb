# Grid Search

In this notebook I'll approch Grid Search, a method for finding the values of the important parameters of a model. Grid Search is simple. It tries all possible combinations of predefined values for each parameter. Here I'll focus on Grid Search but there are other methods for hyperparameter tuning such as (1) Genetic Algorithm, (2) Random Search and, (3) Bayesian optimization.

As I discussed earlier, SVMs are very sensitive to the (1) scale of the dataset and (2) the parameters (specially C and gamma). Therefore I'll use SVC (SVM for classification) for hyperparameter tuning.

*C* is the regularization parameter. A small C means that individual data points have little influence on the decision boundary (low complex model).<br>
*gamma* constrols the width of the Gaussian kernel. A small gamma means that more points will be considered close by.

## Simple Grid Search

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0)

best_score = 0
for gamma in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)
        score = svm.score(X_test, y_test)
        
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}

print("Best score {:.2f}".format(best_score))
print("Best paramaters {}".format(best_parameters))       
```

Okay.. Hopefully you got the idea of Grid Search, but 0.97 is an overly optimistic result because we used the test set to find the best parameters. We shouldn't use the test set for hyperparameter tuning, so we need a portion of the data to validate our parameters before we test it. Such as depicted below.

```python
import mglearn
mglearn.plots.plot_threefold_split()
```

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()

X_train_val, X_test, y_train_val, y_test = train_test_split(iris.data, iris.target, random_state=0)
X_train, X_validation, y_train, y_validation = train_test_split(X_train_val, y_train_val, random_state=1)


best_score = 0
for gamma in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)
        score = svm.score(X_validation, y_validation)
        
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}

            
svm = SVC(**best_parameters)
svm.fit(X_train_val, y_train_val)

test_score = svm.score(X_test, y_test)
print("Best score on validation set: {:.2f}".format(best_score))
print("Best parameters: ", best_parameters)
print("Test set score with best parameters: {:.2f}".format(test_score))    
```

The distinction between the training set, validation set and test set is fundamentally important to applying machine learning methods in practice.

It is a good practice to do all exploratory analysis and model selection using the combination of a training and a validation set, and reserve the test set for a final evaluation.

## Grid-Search with Cross-Validation

Simple Grid-Search inherits the problems of train-test split. So, our results are very dependent of the selected data. We can solve this addind cross-validation to Grid-search.

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score

iris = load_iris()

X_train_val, X_test, y_train_val, y_test = train_test_split(iris.data, iris.target, random_state=0)

best_score = 0
for gamma in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100, 1000]:
        svm = SVC(gamma=gamma, C=C)
        score = np.mean(cross_val_score(svm, X_train_val, y_train_val, cv=5))
        
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}

            
svm = SVC(**best_parameters)
svm.fit(X_train_val, y_train_val)
test_score = svm.score(X_test, y_test)

print("Best score on validation set: {:.2f}".format(best_score))
print("Best parameters: ", best_parameters)
print("Test set score with best parameters: {:.2f}".format(test_score))    
```

The main downside here is the number of models we need to evaluate our model. In this case, we need (7*7) * 5 = 245 models.

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], 'gamma': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}

grid_search = GridSearchCV(SVC(), param_grid, cv=5)

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=0)

grid_search.fit(X_train, y_train)

print("Test set score: {:.2f}".format(grid_search.score(X_test, y_test)))
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python
print("Best estimator:\n{}".format(grid_search.best_estimator_))
```

Let's see all Grid-Search results

```python
import pandas as pd

results = pd.DataFrame(grid_search.cv_results_)

display(results.head())
```

The dataframe contains too much information. Let's visualize it with a Heatmap.

```python
scores = np.array(results.mean_test_score).reshape(7, 7)

mglearn.tools.heatmap(scores, xlabel='gamma', xticklabels=param_grid['gamma'], ylabel='C', yticklabels=param_grid['C'], cmap="viridis");
```

And finally see some cases when the ranges of the parameters are not ideal

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(13, 5))

param_grid_linear = {'C': np.linspace(1, 2, 6), 'gamma': np.linspace(1, 2, 6)}
param_grid_one_log = {'C': np.linspace(1, 2, 6), 'gamma': np.logspace(-3, 2, 6)}
param_grid_range = {'C': np.logspace(-3, 2, 6), 'gamma': np.logspace(-7, -2, 6)}

for param_grid, ax in zip([param_grid_linear, param_grid_one_log, param_grid_range], axes):
    grid_search = GridSearchCV(SVC(), param_grid, cv=5)
    grid_search.fit(X_train, y_train)
    scores = grid_search.cv_results_['mean_test_score'].reshape(6, 6)
 
    scores_image = mglearn.tools.heatmap(scores, xlabel='gamma', ylabel='C', xticklabels=param_grid['gamma'], yticklabels=param_grid['C'], cmap="viridis", ax=ax)

plt.colorbar(scores_image, ax=axes.tolist());
```

### Search over spaces that are not grids

We can also use GridSearchCV with conditional parameters. For example, if kernel = 'linear', *gamma* is not used, but if kernel='rbf', C and gamma are used.

```python
param_grid = [{'kernel': ['rbf'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], 'gamma': [0.001, 0.01, 0.1, 1, 10, 100, 1000]},
             {'kernel': ['linear'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}]

grid_search = GridSearchCV(SVC(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

print("Test set score: {:.2f}".format(grid_search.score(X_test, y_test)))
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python
results = pd.DataFrame(grid_search.cv_results_)
display(results.T)
```

## Nested cross-validation

Up to now, we splitted our data in 3 fixed parts (train, test, validation). This can make ours results unstable and dependent of this single split.

To solve this problem we can apply the nested cross-validation in which we run a cross-validation for differents train, tests splits. To implement this, we can simply pass GridSearchCV as the model for the cross_val_score function.

```python
scores = cross_val_score(GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1), iris.data, iris.target, cv=5)

print("Cross-validation scores: ", scores)
print("Mean cross-validation score: ", scores.mean())
```

As you may expect, this is a very expansive procedure.
