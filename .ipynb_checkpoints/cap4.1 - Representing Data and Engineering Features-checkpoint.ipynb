# Representing Data and Engineering Features


So far, we've dealed with countinuous features. But there are other types of features such as categorical features.
The type of a feature defines the scope of operations we can perform on it. For example, we cannot compute the mean
and the median of a categorical feature, but we can compute the mode.

Type of data is one of the first topics in statistics' classes. We can deal with:

(1) Numerical data

    (1.1) Discrete data (e.g. [2,4,6,8,...] set of even numbers, [1.0 ,1.5, 2.0] a finite set)
    (1.2) Continuous data (e.g. [0, inf[ kelvin scale)

(2) Categorical data

    (2.1) Nominal data (e.g. [male, female])
    (2.2) Ordinal data (e.g. [low, medium, high])

Regardless the type of the data, we want to represent our data as well as possible, so our ML algorithm can learn without
additional problems. How to represent our data best for a particular application is known as *feature engineering*. And, obviously, thats import because it affects the ML's results.


## Categorical Variables


Let's see an example with the adult dataset in which the aim is to predict if an adult has an income of over $50k or under $50k.

### Load dataset and show the first 5 rows. 

```python
import os
import mglearn
import pandas as pd

adult_path = os.path.join(mglearn.datasets.DATA_PATH, "adult.data")

data = pd.read_csv(
 adult_path, header=None, index_col=False,
 names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race',
        'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'])

data = data[['age', 'workclass', 'education', 'gender', 'hours-per-week', 'occupation', 'income']]

display(data.head())
```

```python
data.dtypes
```

As you can see, workclass, education, gender, income and occupation are categorical features. Therefore, it makes no sense to
use their values in a Linear Logistic, for example. In this case, we need numbers not names.

## One-Hot-Encoding (Dummy Variables)

One-Hot-Encoding (or Dummy Variables) is a technique to represent categorical features as discrete features.
The idea here is to replace each categorical feature with one or more binary features in which 1 represents the presence
of categorical value and 0 the absence. For example, "gender" would generate two new features "gen_Male" and "gen_Female",
these two features admits only 0 and 1. If we have 1 for "gen_Male" it indicates the presence of "Male" value and "gen_Female"
must be 0 for this row.

### Application of One-Hot-Encoding

```python
print("Original features:\n", list(data.columns), "\n")
data_dummies = pd.get_dummies(data)
print("Features after get_dummies:\n", list(data_dummies.columns))
```

### Separating features (dependent (y) and independent variables (X))

```python
features = data_dummies.loc[:, 'age':'occupation_ Transport-moving']


X = features.values
y = data_dummies['income_ >50K'].values
#y = data['income'].apply(lambda x: -1 if x == ' <=50K' else 1).values #Alternative

print("X.shape: {} y.shape: {}".format(X.shape, y.shape))
```

### Application of Machine learning algorithm

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

logreg = LogisticRegression(max_iter=1200)
logreg.fit(X_train, y_train)

print("Train score: {:.4f} Test score: {:.4f}".format(logreg.score(X_train, y_train), logreg.score(X_test, y_test)))
```

## Binning, Discretization, Linear Models and Trees

The best way to represent data depends not only on the semantics of the data, but also on the kind of model we're using.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

X, y = mglearn.datasets.make_wave(n_samples=100)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1, 1)

reg = DecisionTreeRegressor(min_samples_split=3).fit(X, y)
plt.plot(line, reg.predict(line), label="Decision Tree")

print("Decision Tree - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

reg = LinearRegression().fit(X, y)
plt.plot(line, reg.predict(line), label="Linear Regression")

print("Linear Regression - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel('Regression output')
plt.xlabel('Input feature')

plt.legend(loc='best');
```

### Split our data into N=10 bins

```python
bins = np.linspace(-3, 3, 11)
print("bins: {}".format(bins))
```

### Distribute data among the bins created above

```python
which_bin = np.digitize(X, bins=bins)
print("\nData points:\n", X[:5])
print("\nBin membership for data points:\n", which_bin[:5])
```

### Using OneHotEncoding to produce 10 new features from binned data

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoder.fit(which_bin)

X_binned = encoder.transform(which_bin)

print(X_binned[:5])
```

### Application of ML algorithms on binned data

```python
plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

#X, y = mglearn.datasets.make_wave(n_samples=100)
#line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1, 1)

line_binned = encoder.transform(np.digitize(line, bins=bins))

X_train, X_test, y_train, y_test = train_test_split(X_binned, y, random_state=0)

reg = DecisionTreeRegressor(min_samples_split=3).fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label="Decision Tree (binned)")
print("Decision Tree - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

reg = LinearRegression().fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label="Linear Regression (binned)", linestyle='dashed')
print("Linear Regression - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

plt.plot(X[:, 0], y, 'o', c='k')
plt.vlines(bins, -3, 3, linewidth=1, alpha=.2)
plt.ylabel('Regression output')
plt.xlabel('Input feature')

plt.legend(loc='best');
```

If we have good reasons to use a Linear Model and we have to deal with features of nonlinear relations with the output,
we can use binning to increase modeling power.

## Interactions and Polynomials

Adding *interaction features* and *polynomial features* to our data can  enrich our feature representation.
Linear models can learn offsets and slopes. We can enrich our previous data representation by inserting the original X feature to our data.

```python
X_combined = np.hstack([X, X_binned])
print('Original X\'s shape:', X.shape, '\nBinned X\'s shape:', X_binned.shape, '\nCombined X\'s shape:', X_combined.shape)
```

```python
plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

X_train, X_test, y_train, y_test = train_test_split(X_combined, y, random_state=0)

reg = LinearRegression().fit(X_combined, y)
line_combined = np.hstack([line, line_binned])

plt.plot(line, reg.predict(line_combined), label='linear regression combined')
print("Linear Regression - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')

plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.plot(X[:, 0], y, 'o', c='k');
```

Learned sloped is downward and shared across all the bins because we have a single x-axis (continuous) feature.
We should have a singular slope for each bin. We can achieve this by separating the x-axis.

```python
X_product = np.hstack([X_binned, X * X_binned])
print(X_product.shape)
```

Now we have the binned features and one x-axis (continuous) for each binned feature.

```python
plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

reg = LinearRegression().fit(X_product, y)

line_product = np.hstack([line_binned, line * line_binned])

X_train, X_test, y_train, y_test = train_test_split(X_product, y, random_state=0)

plt.plot(line, reg.predict(line_product), label='linear regression product')
print("Linear Regression - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')
    
plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.legend(loc="best");
```

Data binning is a way of expand original features. Another way is to produce new polynomial features from the original dataset.

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=10, include_bias=False)
poly.fit(X)
X_poly = poly.transform(X)
```

```python
print("X_poly.shape: {}".format(X_poly.shape))
print("\nEntries of X:\n{}".format(X[:5]))
print("\nEntries of X_poly:\n{}".format(X_poly[:5]))
print("\nPolynomial feature names:\n{}".format(poly.get_feature_names()))
```

As you can see, we added Polynomial Features to our dataset which is nothing but each features raised to some number. With polynomial featuer, we can increase the power of the linear regression classifier. Below, we show the results of linear regression application in a dataset with these polynomial features.

```python
plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

reg = LinearRegression().fit(X_poly, y)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state=0)
print("Linear Regression - Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))

line_poly = poly.transform(line)

plt.plot(line, reg.predict(line_poly), label='polynomial linear regression')
plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.legend(loc="best");
```

Now, linear regression produces a very smooth fit. However, a kernel SVM model can produce the same result without any transformation (see below).

```python
from sklearn.svm import SVR

plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

for gamma in [1, 10, 100]:
    svr = SVR(gamma=gamma).fit(X, y)
    plt.plot(line, svr.predict(line), label='SVR gamma={}'.format(gamma))
    print("SVM (gamma: {:.2f}) - Test set R^2: {:.2f}".format(gamma, SVR(gamma=gamma).fit(X_train, y_train).score(X_test, y_test)))
    
plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.legend(loc="best");
```

### Application of Interactions and Polynomial Features to the Boston House Dataset

```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

boston = load_boston()

X_train, X_test, y_train, y_test = train_test_split(boston.data, boston.target, random_state=0)

# rescale data
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
poly = PolynomialFeatures(degree=2).fit(X_train_scaled)

X_train_poly = poly.transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

print("X_train.shape: {}".format(X_train.shape))
print("X_train_poly.shape: {}".format(X_train_poly.shape))
print("\nPolynomial feature names:\n{}".format(poly.get_feature_names()))
```

### Ridge's Results

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train_scaled, y_train)
print("Score without interactions: {:.3f}".format(ridge.score(X_test_scaled, y_test)))

ridge = Ridge().fit(X_train_poly, y_train)
print("Score with interactions: {:.3f}".format(ridge.score(X_test_poly, y_test)))
```

### RandomForest's Results

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100).fit(X_train_scaled, y_train)
print("Score without interactions: {:.3f}".format(rf.score(X_test_scaled, y_test)))

rf = RandomForestRegressor(n_estimators=100).fit(X_train_poly, y_train)
print("Score with interactions: {:.3f}".format(rf.score(X_test_poly, y_test)))
```

As you can notice, RandomForest outperforms RidgeRegressor with and without interactions. Actually, RandomForest produces better results without any transformation, meanwhile Ridge benefits a lot from transformations applied to the data.

##  Univariate Nonlinear Transformations

As we saw before, squared or cubed features can help linear models for regression. There are other transformations such as log, exp and sin that have been proven to also be useful.

Tree-based models: care about the ordering of the features
Linear Models and Neural Networks: are tied to the scale and distribution of each feature.

If there is a nonlinear relationship between the feature and the target, that becomes hard to model - particularly in regression.

*log* and *exp* can help by adjusting the relative scales in the data. Meanwhile, *sin* and *cos* come in handy when dealing with periodic data.

Below is an example:

```python
rnd = np.random.RandomState(0)
X_org = rnd.normal(size=(1000, 3))

w = rnd.normal(size=3)
X = rnd.poisson(10 * np.exp(X_org))
y = np.dot(X_org, w)
```

```python
print("Number of feature appearances:\n{}".format(np.bincount(X[:, 0])))
```

```python
from sklearn.linear_model import Ridge

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
score = Ridge().fit(X_train, y_train).score(X_test, y_test)

print("Test score: {:.3f}".format(score))
```

```python
plt.figure(figsize=(8,6), dpi= 100, facecolor='w', edgecolor='k')

bins = np.bincount(X[:, 0])
plt.bar(x=range(len(bins)), height=bins, color='w', edgecolor='black')

plt.ylabel("Number of appearances")
plt.xlabel("Value");
```

```python
X_train_log = np.log(X_train + 1)
X_test_log = np.log(X_test + 1)

plt.hist(X_train_log[:, 0], bins=25, color='gray', edgecolor='black')
plt.ylabel("Number of appearances")
plt.xlabel("Value");
```

```python
score = Ridge().fit(X_train_log, y_train).score(X_test_log, y_test)

print("Test score: {:.3f}".format(score))
```
